```python
import pandas as pd
import matplotlib.pyplot as plt

Inception_path = '/root/7.Proposed Version/PretrainedComparison/Inception_Pretrained.csv'

VGG_Path = '/root/7.Proposed Version/PretrainedComparison/VGG16_Pretrained.csv'

ResNet_Path = '/root/7.Proposed Version/PretrainedComparison/ResNet50V2.csv'

My_Path = '/root/7.Proposed Version/PretrainedComparison/MYMODEL.csv'

# 
inception_data = pd.read_csv(Inception_path)
vgg_data = pd.read_csv(VGG_Path)
resnet_data = pd.read_csv(ResNet_Path)
my_model_data = pd.read_csv(My_Path)

#  ResNet 
resnet_data.rename(columns=lambda x: x.replace('_1', ''), inplace=True)

# 
def plot_metric_comparison_all_models(metric):
    plt.figure(figsize=(12, 6))

    # 
    plt.plot(vgg_data['epoch'], vgg_data[metric], label='VGG16', marker='o')
    plt.plot(inception_data['epoch'], inception_data[metric], label='Inception', marker='^')
    plt.plot(resnet_data['epoch'], resnet_data[metric], label='ResNet50V2', marker='s')
    plt.plot(my_model_data['epoch'], my_model_data[metric], label='ProposedModel', marker='x')

    plt.title(f'Comparison of {metric} Across Models')
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.show()

# 
metrics = ['accuracy', 'loss','f1_score', 'precision', 'recall', 'auc',
          'val_accuracy', 'val_loss','val_f1_score', 'val_precision', 'val_recall','val_auc']
for metric in metrics:
    plot_metric_comparison_all_models(metric)
```
