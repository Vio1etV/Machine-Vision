```python
 
import os,random,shutil
 
```

```python

def moveFile(fileDir, tarDir):
    pathDir = os.listdir(fileDir)  # 取图片的原始路径
    filenumber = len(pathDir)
    rate = 0.85  # 自定义抽取图片的比例，比方说100张抽10张，那就是0.1
    picknumber = int(filenumber * rate)  # 按照rate比例从文件夹中取一定数量图片
    sample = random.sample(pathDir, picknumber)  # 随机选取picknumber数量的样本图片
    print(sample)
    print(len(sample))
    for name in sample:
        src = os.path.join(fileDir, name)
        dst = os.path.join(tarDir, name)
        shutil.move(src, dst)
 
 
if __name__ == '__main__':
    # 源图片文件夹路径
    fileDir = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Corsework\Machine Vision\Coursework\DR Datasets\colored_images\Diabetic' 
     # 移动到新的文件夹路径
    tarDir = r'C:\Users\MSI-PC\Desktop\Level 6 curriculum\Corsework\Machine Vision\Coursework\DR Datasets\DRD_Split\train\diabetic'
    moveFile(fileDir, tarDir)
```
